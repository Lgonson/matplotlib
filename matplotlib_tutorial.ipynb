```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
# Plot y versus x as lines.
x = [0, 1,2,3, 4]
y = [0, 2,4,6, 8]

plt.plot(x,y, label='y = 2x', color = 'red', linewidth=1, marker='.', markersize=10, markeredgecolor='blue', linestyle='--')

plt.title('First Graph', fontdict={'fontname':'Comic Sans MS'})
plt.xlabel('X Axis', fontdict={'fontname':'Comic Sans MS'})
plt.ylabel('Y Axis', fontdict={'fontname':'Comic Sans MS'})

#ticks (points in the plot)
plt.xticks([0, 1,2,3, 4])
plt.yticks([0, 2,4,6, 8])

plt.legend() #if we have 
plt.show()
```

```python
# Plot y versus x as lines.
x = [0, 1,2,3, 4]
y = [0, 2,4,6, 8]

#Resize plot
plt.figure(figsize=(5,3), dpi=100)

plt.title('Third Graph', fontdict={'fontname':'Comic Sans MS'})
plt.xlabel('X Axis', fontdict={'fontname':'Comic Sans MS'})
plt.ylabel('Y Axis', fontdict={'fontname':'Comic Sans MS'})

#First Line
plt.plot(x,y, 'bo-', label='2X')
#Second Line
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:6], x2[:6]**2, 'r', label='x2^2')
plt.plot(x2[5:], x2[5:]**2, 'r--')

plt.savefig('my_plot.png', dpi=300)
plt.legend() #if we have 
plt.show()
```

# Bar chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)
bars[0].set_hatch('/')
bars[1].set_hatch('|')
bars[2].set_hatch('+')

plt.figure(figsize=(4,4))
plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/', '*', 'O']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.figure(figsize=(4,4))
plt.show()
```
