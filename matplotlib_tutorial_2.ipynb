```python
plt.figure(figsize=(7,5))
#markers: https://matplotlib.org/3.5.0/api/markers_api.html

for country in df:
    if country != 'Year':
        plt.plot(df.Year, df[country], marker='.', label=[country])

plt.title('Gas Prices in USD')
plt.xlabel('Year')
plt.ylabel('USD')

plt.xticks(df.Year[::3])

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(7,5))


countries_to_look = ['France', 'Germany', 'USA', 'Mexico', 'Japan']
for country in countries_to_look:
    if country != 'Year':
        plt.plot(df.Year, df[country], marker='.', label=[country])

plt.title('Gas Prices in USD', fontdict={'fontweight':'bold', 'fontsize':'22'})
plt.xlabel('Year')
plt.ylabel('Gas Price in USD')

plt.xticks(df.Year[::3].tolist()+[2011])

plt.legend()
plt.savefig('GasPriceInUSD.png', dpi=300)

plt.show()
```

# Load Fifa DF

```python
df = pd.read_csv(r'C:\Users\lgons\workspace\Python\Matplotlib_tutorial\fifa_data.csv')
```

# Histogram

```python
bins=[30,40,50,60,70,80,90,100]

#Google color picker: https://www.google.com/search?client=firefox-b-d&q=google+color+picker
plt.hist(df.Overall, bins=bins, color='#abcdef')
plt.xlabel('Skill level')
plt.ylabel('Number of Players')
plt.title('Distribution of Players Skills in Fifa 2018')

plt.show()
```

# Pie Chart

```python
right = df.loc[df['Preferred Foot'] == 'Right'].count().iloc[0]
left = df.loc[df['Preferred Foot'] == 'Left'].count().iloc[0]

labels = ['Left', 'Right']
colors = ['#b8dae3', '#b6bfc2']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')
plt.title('Foot Preferences of Fifa Players')
plt.show()
```

```python


df.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in df.Weight]

plt.style.use('ggplot')
#plot styles: https://matplotlib.org/3.1.0/gallery/style_sheets/style_sheets_reference.html

light = df.loc[df.Weight < 125].count().iloc[0]
light_medium = df.loc[(df.Weight > 125) & (df.Weight < 150)].count().iloc[0]
medium = df.loc[(df.Weight > 150) & (df.Weight < 175)].count().iloc[0]
medium_heavy = df.loc[(df.Weight > 175) & (df.Weight < 200)].count().iloc[0]
heavy = df.loc[df.Weight > 200].count().iloc[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = [0.4, 0.2, 0, 0, 0.4]

plt.title('Weight distribution of FIFA players (in lbs)')
plt.pie(weights, labels = labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)
plt.show()
```

```python
plt.style.use('default')
barcelona = df.loc[df.Club == 'FC Barcelona']['Overall']
madrid = df.loc[df.Club == 'Real Madrid']['Overall']
river = df.loc[df.Club == 'River Plate']['Overall']

tick_labels = ['FC Barcelona', 'Real Madrid', 'River Plate']
boxes = plt.boxplot([barcelona, madrid, river], tick_labels = tick_labels, patch_artist=True, medianprops={'linewidth':'2'})

for box in boxes['boxes']:
    #Set edge color
    box.set(color='#4286f4', linewidth=2)
    #Change fill color (to change this fill color, we should add the patch_artist=True)
    box.set(facecolor='#e0e0e0')

plt.title('Professional Futbol Team Comparisson')
plt.ylabel('Fifa Overall Rating')
plt.show()
```
